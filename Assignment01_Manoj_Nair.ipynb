### Qn. 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

##### #1 Scatter plot for "Petal Length vs Petal width" & "Sepal Length vs Sepal width"

```python
plt.style.use('seaborn-deep')
plt.figure(figsize=(16,6))
sns.scatterplot(data=df,x ='PL', y="PW", hue='Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Petal Width of Iris')
```

```python
plt.style.use('seaborn-deep')
plt.figure(figsize=(16,6))
sns.scatterplot(data=df,x ='SL', y="SW", hue='Classification')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width of Iris')
```

##### #2 Box plot for "Sepal Length vs Sepal Width" & "Petal Length vs Petal width"

```python
plt.figure(figsize=(20,8))
sns.boxplot(x='SL',y='SW',hue='Classification',data=df)
plt.title('Sepal Length vs Sepal Width',size=20)
plt.figure(figsize=(20,10))
sns.boxplot(x='PL',y='PW',hue='Classification',data=df)
plt.title('Petal Length vs Petal Width',size=20)
```

##### #3 Swarm plot of "Sepal Width" & "Petal Width"

```python
plt.figure(figsize=(15,8))
sns.swarmplot(x='Classification',y='SW',data=df,hue='Classification')
plt.title('Sepal Width of different Classification')
```

```python
plt.figure(figsize=(15,8))
sns.swarmplot(x='Classification',y='PW',data=df,hue='Classification')
plt.title('Petal Width of different Classification')
```

##### #4: Histogram for "Sepal Length" & "Petal Length"

```python
plt.figure(figsize=(10,7))
plt.hist(df['SL'],bins=20,color='black')
plt.title('Sepal Length in cm')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(10,7))
plt.hist(df['PL'],bins=20,color='indigo')
plt.title('Petal Length in cm')
plt.xlabel('Petal Length')
plt.ylabel('Count')
```
